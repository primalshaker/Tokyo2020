```python
###tokyo2020 GroupBy countries
def MedalStat(x):
    gold = x[x["medal"] == "gold"]["medal"].count()
    silver = x[x["medal"] == "silver"]["medal"].count()
    bronze = x[x["medal"] == "bronze"]["medal"].count()
    total = gold + silver + bronze
    total_weighted = gold*3 + silver*2 + bronze
    return pd.Series([gold, silver, bronze, total, total_weighted], 
                     index=["Gold", "Silver", "Bronze", "Total", "TotalWeighted"])

expected_df = tokyo2020.groupby(["country"]).apply(MedalStat)

# order by various ways
expected_df.sort_values(['Gold', 'Silver', 'Bronze'], ascending=False).head(60) # By Gold, Silver, Bronze
expected_df.sort_values(['Total'], ascending=False).head(60) # By Total
expected_df.sort_values(['TotalWeighted'], ascending=False).head(60) # By TotalWeighted
```

```python
expected_df["rank_Total_min"]=expected_df["Total"].rank(method="min", ascending=False)
expected_df["rank_Total_weighted"]=expected_df["TotalWeighted"].rank(method="min", ascending=False)
expected_df.sort_values(['rank_Total_min'], ascending=True).head(60) # By Total
```
